```python
# It is important not to use inline but notebook instead
%matplotlib notebook

from math import sin,cos
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.arange(-10,10,1)
y = [3*cos(t) for t in x]
y = np.array(y)

fig = plt.figure()
plt.xlim(-10, 10)
plt.ylim(-10, 10)
ax, = plt.plot([], [],'g*',color='g',linestyle='dashed',linewidth=1 )
plt.grid()

def animate(i):
    ax.set_data(x[:i], y[:i])
    return ax

ani = FuncAnimation(fig, animate, frames=20, interval=200)

# Saving the animation in gif form factor
#from matplotlib.animation import FFMpegWriter
#writer = FFMpegWriter(fps=15, metadata=dict(artist='Me'), bitrate=1800)
#ani.save('myAnimation.gif', writer='imagemagick', fps=20)
```
